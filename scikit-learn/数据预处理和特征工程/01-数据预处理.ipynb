```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
pd.DataFrame(data)
```

```python
scacler = preprocessing.MinMaxScaler().fit(data)
result = scacler.transform(data)
result
```

```python
result_ = scacler.inverse_transform(result)
result_
```

```python
# MinMaxScaler 调整feature_range参数 使数据缩放到指定的区间
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = preprocessing.MinMaxScaler(feature_range=(5, 10))
result = scaler.fit_transform(data)
result
```

```python
# 数据标准化
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
scaler = preprocessing.StandardScaler()
result = scaler.fit_transform(data)
result
```

```python
scaler.inverse_transform(result)
```

```python
# 预处理分类型特征
data = pd.read_csv('./data/Narrativedata.csv', index_col=0)
data.head()
```

```python
y = data.iloc[:, -1]
le = preprocessing.LabelEncoder().fit(y)
result = le.transform(y)
result
```

```python
le.classes_
```

```python
data.iloc[:, -1] = result
data.head()
```

```python
# 处理缺失值
data.loc[:, 'Age'] = data.loc[:, 'Age'].fillna(data.loc[:, 'Age'].mean())
```

```python
data.info()
```

```python
data.dropna(axis=0, inplace=True)
```

```python
# OrdinalEncoder 特征专用 能够将分类特征转化为分类数值
# 如果特征值中有Nan会报错
data_ = data.copy()
data_.head()
```

```python
 preprocessing.OrdinalEncoder().fit(data_.iloc[:, 1:-1]).categories_
```

```python
data_.iloc[:, 1:-1] = preprocessing.OrdinalEncoder().fit_transform(data_.iloc[:, 1:-1])
data_.head()
```

```python
# 独热编码 (哑变量)
X = data.iloc[:, 1:-1]
enc = preprocessing.OneHotEncoder().fit(X)
result = enc.transform(X).toarray()
result
```

```python
# 下面的特征和上面的矩阵列一一对应
enc.categories_
```

```python
enc.get_feature_names()
```

```python
# 然后对数据进行跨行合并
newdata = pd.concat([data, pd.DataFrame(result)], axis=1)
newdata.head()
```

```python
newdata.drop(['Sex', 'Embarked'], axis=1, inplace=True)
```

```python
newdata.head()
```

```python
newdata.columns = ['Age', 'Survived', 'Female', 'Male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']
newdata.head()
```

```python
# preprocessing.Binarizer  二值化
# preprocessing.Binarizer(threshold=10)
# 分箱操作 preprocessing.KBinsDiscretizer  因为要处理特征 所以必须是二维数组
X = data.iloc[:, 0].values.reshape(-1, 1)
X
```

```python
# n_bins 分箱操作
# encode： ordinal，onehot，onehot-dense  
# ordinal 其实就相当于OrdinalEncoder
# onehot 哑变量后 返回一个稀疏矩阵
# onehot-dense 哑变量后返回一个密集数组
est = preprocessing.KBinsDiscretizer(n_bins=3, encode='onehot', strategy='uniform')
est.fit_transform(X).toarray()
```
