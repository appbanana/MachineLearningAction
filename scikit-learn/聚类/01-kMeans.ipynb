```python
from sklearn import datasets, cluster, metrics
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
X, y = datasets.make_blobs(n_samples=1000, n_features=2, centers=4, random_state=66)
X.shape
```

```python
fig, ax1 = plt.subplots(1)
ax1.scatter(X[:, 0], X[:, 1], marker='o', s=8)
plt.show()
```

```python
colors = ['red', 'pink', 'orange', 'gray']
fig, ax1 = plt.subplots(1)
for i in range(4):
    ax1.scatter(X[y==i, 0], X[y==i, 1], marker='o', s=8, color=colors[i])
plt.show()
```

```python
n_clusters = 3
cluster_ = cluster.KMeans(n_clusters=n_clusters, random_state=66).fit(X)
```

```python
y_pred = cluster_.labels_
y_pred
```

```python
pre = cluster_.fit_predict(X)
pre == y_pred
```

```python
# 实际迭代的次数
cluster_.n_iter_
```

```python
# 每个样本点距离他们最近簇心的均方距离，又叫簇内平方和
cluster_.inertia_
```

```python
# 收敛的质心
cluster_.cluster_centers_
```

```python
# 返回的是所有样本轮廓系数的均值
metrics.silhouette_score(X, y_pred)
```

```python
# 返回数据集中每个样本的轮廓系数
metrics.silhouette_samples(X, y_pred)
```

```python
metrics.calinski_harabaz_score(X, y_pred)
```

```python
# 基于轮廓系数来选择n_clusters
n_clusters = 4
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(18, 7)
ax1.set_xlim([-0.1, 1])
clusterr = cluster.KMeans(n_clusters=n_clusters, random_state=66).fit(X)
cluster_label = clusterr.labels_
silhouette_avg = metrics.silhouette_score(X, cluster_label)
print('簇数为：', n_clusters, '轮廓系数平均值为：', silhouette_avg)

sample_silhouette_value = metrics.silhouette_samples(X, cluster_label)
y_lower = 10
for i in range(n_clusters):
    ith_cluster_silhouette_value = sample_silhouette_value[cluster_label == i]
    ith_cluster_silhouette_value.sort()
    size_cluster_i = ith_cluster_silhouette_value.shape[0]
    y_upper = y_lower + size_cluster_i
    color = cm.nipy_spectral(float(i) / n_clusters)
    ax1.fill_betweenx(np.arange(y_lower, y_upper), ith_cluster_silhouette_value, facecolor=color, alpha=0.7)
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10
ax1.set_title("The silhouette plot for the various clusters.")
ax1.set_xlabel("The silhouette coefficient values")
ax1.set_ylabel("Cluster label")

clolors = cm.nipy_spectral(cluster_label.astype(float) / n_clusters)
ax2.scatter(X[:, 0], X[:, 1], marker='o', s=8, c=y)
centers = clusterr.cluster_centers_
ax2.scatter(centers[:, 0], centers[:, 1], marker='x', c='red', alpha=1, s=200)

ax2.set_title("The visualization of the clustered data.")
ax2.set_xlabel("Feature space for the 1st feature")
ax2.set_ylabel("Feature space for the 2nd feature")
plt.suptitle(("Silhouette analysis for KMeans clustering on sample data"
              "with n_clusters = %d" % n_clusters),
             fontsize=14, fontweight='bold')

plt.show()
```
