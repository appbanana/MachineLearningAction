```python
# 98. 验证二叉搜索树
# https://leetcode-cn.com/problems/validate-binary-search-tree/
```

```python
class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
        

class Solution:
    def __init__(self):
        self.last = None
        
    def isValidBST1(self, root: TreeNode) -> bool:
        """
        思路: 中序遍历  因为二叉搜索树的中序遍历是升序或者降序排列
        """
        if root is None: return True
        
        if not self.isValidBST(root.left): 
            return False
        if self.last and root.val <= self.last:
            return False
        last = root.val
        
        if not self.isValidBST(root.right):
            return False
        return True
        
    def isValidBST2(self, root: TreeNode) -> bool:
        """
        中序遍历 后一个值比前一个值大
        """
        if root is None: 
            return True
        stack = []
        cur = root
        last_val =  float('-inf')
        while cur or len(stack):
            if cur is not None:
                stack.append(cur)
                cur = cur.left
            else:
                node = stack.pop()
                if node.val <= last_val:
                    return False
                cur = node.right
                last_val = node.val
        return True
    
    def isValidBST2(self, root: TreeNode) -> bool:
        """
        层序遍历 
        """
        if root is None: 
            return True
        node_queue = []
        min_queue = []
        max_queue = []
        
        node_queue.append(root)
        min_queue.append(None)
        max_queue.append(None)
        
        while len(node_queue):
            node = node_queue.pop(0)
            min_val = min_queue.pop(0)
            max_val = max_queue.pop(0)
            # 比下届还小 直接返回false
            if min_val and node.val <= min_val: 
                return False
            # 比上届大 直接返回true
            if max_val and node.val >= max_val:
                return False
            if node.left is not None:
                node_queue.append(node.left)
                min_queue.append(min_val)
                max_queue.append(node.val)
            if node.right is not None:
                node_queue.append(node.right)
                min_queue.append(node.val)
                max_queue.append(max_val)
            
        return True
```
