```python
# 102. 二叉树的层次遍历
# https://leetcode-cn.com/problems/binary-tree-level-order-traversal/

class TreeNode:
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None

class Solution:

    # 尾递归来实现 层次遍历 
    # 执行用时 : 52 ms, 在Binary Tree Postorder Traversal的Python3提交中击败了85.66% 的用户
    # 内存消耗 : 12.9 MB, 在Binary Tree Postorder Traversal的Python3提交中击败了99.14% 的用户
    def levelOrder(self, root: TreeNode) -> list:
        order_list = []
        self.level(root, order_list)
        return order_list
    
    def level(self, node: TreeNode, order_list:list): 
        if node is None: return 
        # 这个递归我暂时还没想好怎么搞
        pass
#         order_list.append(node.val)
#         self.level(node.left, order_list)
#         self.level(node.right, order_list) 
    
    # 迭代来实现 后序 左 右 根 输出这种格式的[[1], [2, 3], [4, 5, 6, 7]]
    # 执行用时 : 64 ms, 在Binary Tree Level Order Traversal的Python3提交中击败了46.54% 的用户
    # 内存消耗 : 13 MB, 在Binary Tree Level Order Traversal的Python3提交中击败了99.59% 的用户
    def levelOrder1(self, root: TreeNode) -> list:
        if root is None: return []
        stack = [root]
        order_list = []
        inner_list = []
        level_size = 1
        while len(stack):
            node = stack.pop(0)
            inner_list.append(node.val)
            level_size = level_size -1
            if node.left:
                stack.append(node.left)
            if node.right:
                stack.append(node.right)
            if level_size == 0:
                level_size = len(stack)
                order_list.append(inner_list)
                # 这不可能用clear 这涉及到深层拷贝
                inner_list = []

        return order_list
    
    # 迭代来实现 后序 左 右 根 输出这种格式的[1, 2, 3, 4, 5]
    def levelOrder2(self, root: TreeNode) -> list:
        if root is None: return []
        stack = [root]
        order_list = []
        while len(stack):
            node = stack.pop(0)
            order_list.append(node.val)
            if node.left:
                stack.append(node.left)
            if node.right:
                stack.append(node.right)
        return order_list
    
```
