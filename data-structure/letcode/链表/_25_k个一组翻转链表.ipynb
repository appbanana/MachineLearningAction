```python
# 25. k个一组翻转链表 # 当 k = 3 时，应当返回: 3->2->1->4->5
#  https://leetcode-cn.com/problems/reverse-nodes-in-k-group/comments/
class ListNode:
    def __init__(self, x=0):
        self.val = x
        self.next = None

class Solution:    
    def reverseKGroup(self, head: ListNode, k: int) -> ListNode:
        if head is None: return head
        temp = head
        for i in range(k):
            temp = temp.next
            if temp is None: 
                return head  
        last = self.reverseKGroup(temp, k)
        cur = head
        next_node = None
        for i in range(k):
            next_node = cur.next
            cur.next = last
            last = cur
         
        return last
    

        
```

```python
first = ListNode()
first.next = None
last_node = first
for i in range(1, 10):
    node = ListNode(i)
    last_node.next = node
    last_node = node
```
