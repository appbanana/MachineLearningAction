```python
# 237. 删除链表中的节点
#  https://leetcode-cn.com/problems/delete-node-in-a-linked-list/
class ListNode:
    def __init__(self, x=0):
        self.val = x
        self.next = None

class Solution:
    def hasCycle(self, head):
        """
        快慢指针 时间复杂度o(n), 空间复杂度o(1)
        :type head: ListNode
        :rtype: bool
        """
        if head is None or head.next is None:
            return False
        slow = head.next
        quick = head.next
        while quick is not None and quick.next is not None:
            slow = slow.next
            quick = quick.next.next
            if slow == quick:
                return True

        return False
        
```

```python
first = ListNode()
first.next = None
last_node = first
for i in range(1, 10):
    node = ListNode(i)
    last_node.next = node
    if i == 8:
        last_node.next = first
    last_node = node
```
